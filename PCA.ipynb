```python

import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA


X = np.array([[0, 0], [0, 1], [2, 1], [3, 4], [5, 2]])

# Crear una instancia de PCA y ajustarla a los datos
pca = PCA(n_components=2)  # En este caso, reducimos a 2 componentes principales
X_pca = pca.fit_transform(X)

# Mostrar los componentes principales
print("Componentes principales:")
print(X_pca)

# Mostrar la varianza explicada por cada componente principal
print("Varianza explicada por cada componente principal:")
print(pca.explained_variance_ratio_)

# Graficar los datos originales y transformados
plt.figure(figsize=(8, 6))

# Graficar los datos originales
plt.scatter(X[:, 0], X[:, 1], color='blue', label="Datos originales")

# Graficar los datos proyectados en el nuevo espacio PCA
plt.scatter(X_pca[:, 0], X_pca[:, 1], color='red', label="Datos PCA")
plt.xlabel('Componente Principal 1')
plt.ylabel('Componente Principal 2')
plt.title('PCA de los datos')
plt.legend()
plt.grid(True)
plt.show()
```
